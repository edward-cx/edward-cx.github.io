The name of this dataset is "Current_Employee_Names__Salaries__and_Position_Titles". It represents the information about all the current employees of the City of Chicago, which includes their names(object), job titles(object), departments they worked for(object), their working status(full or part time, object) and their payment method(salary or hourly, object). Also, for salary paid employees, there is a annual salary(float64), and for hourly paid employees, there are typical hours they worked(float64) and their hourly rate payment(float64).

```python
import pandas as pd
import bqplot
import numpy as np
import ipywidgets
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('/Users/CHENYX/590DV/Current_Employee_Names__Salaries__and_Position_Titles.csv')
data
```

```python
data.columns
```

```python
data.dtypes
```

```python
salary = data[data['Salary or Hourly']=='Salary']
salary
```

```python
# group the dataset by "Department" and "Job Titles" to find the total annual salary value
total = salary.groupby(['Department','Job Titles'],as_index =False)['Annual Salary'].sum()
total
```

```python
t = total.nlargest(100,'Annual Salary')
t
```

```python
fig1_data = t.pivot(index = 'Department', columns = 'Job Titles', values = 'Annual Salary')
fig1_data
```

```python
dept = fig1_data.index.values.tolist()
dept
```

```python
converse = t.pivot(index = 'Job Titles', columns = 'Department', values = 'Annual Salary')
job = converse.index.values.tolist()
job
```

```python
total2 = salary.groupby(['Department','Job Titles','Name'],as_index =False)['Annual Salary'].sum()
total2
```

```python
# CREATE LABEL - #1
mySelectedLabel = ipywidgets.Label()
```

```python
# CREATE HEATMAP ELEMENTS - #2
# scales
col_sc = bqplot.ColorScale(scheme="Red")
x_sc = bqplot.OrdinalScale()
y_sc = bqplot.OrdinalScale()

# axis
c_ax = bqplot.ColorAxis(scale = col_sc, orientation = 'vertical', side = 'right')
x_ax = bqplot.Axis(scale = x_sc, label = 'Job Titles')
y_ax = bqplot.Axis(scale = y_sc, orientation='vertical', label = 'Department')

# marks
heat_map = bqplot.GridHeatMap(color=fig1_data, 
                              scales ={'color':col_sc, 'row': y_sc, 'column': x_sc}, 
                             interactions={'click':'select'}, # an interaction associated with GHM
                             anchor_style = {'fill':'blue'}) 

# create the grid heat map
fig = bqplot.Figure(marks=[heat_map], axes = [c_ax, x_ax, y_ax])
fig
```

```python
# CREATE SCATTER PLOT ELEMENTS - #3
# scales
x_sch = bqplot.OrdinalScale() # range of data along Z -axis => 0->1
y_sch = bqplot.LinearScale() # frequency, counts

# axis
x_axh = bqplot.Axis(scale = x_sch, label='Name')
y_axh = bqplot.Axis(scale = y_sch, 
                   orientation='vertical',
                   label = 'Annual Salary')

# marks
# a sample dataset used for sample scatter plot
sample =  total2[(total2['Department'] == 'ADMIN HEARNG') & (total2['Job Titles'] == 'ADMINISTRATIVE ASST II')]

hist = bqplot.Scatter(x = sample['Name'], y = sample['Annual Salary'], scales={'x':x_sch, 'y':y_sch})

# create the scatter plot
fig_hist = bqplot.Figure(marks=[hist], axes=[x_axh, y_axh])
fig_hist
```

```python
# LINKING SCATTER PLOT WITH HEATMAP
def on_selected(change):
    if len(change['owner'].selected) == 1: # only selecting 1 element at a time
        i,j = change['owner'].selected[0] # y & x indicies
        v = fig1_data.at[dept[i],job[j]] # x, y indicies, grab data value at this location
        mySelectedLabel.value = 'Annual Salary = ' + str(v) # update widget label with this value
        
        # link the scatter plot
        s = total2[(total2['Department'] == dept[i]) & (total2['Job Titles'] == job[j])]
        hist.x = s['Name']
        hist.y = s['Annual Salary']

# make sure we "observe" for a change in our heatmap (traitlets)
heat_map.observe(on_selected, 'selected')
```

```python
# layout - nested HBox & VBox
figures = ipywidgets.HBox([fig, fig_hist])

# change how big my figures appear on my output
fig.layout.min_width='400px'
fig_hist.layout.min_width='400px'

# dashboard
myDashboard = ipywidgets.VBox([mySelectedLabel, figures])
myDashboard
```

The main purpose of this dashboard is to visually characterize the total annual salary of different jobs in each department and the distribution of the total annual salary of each employee under each job. The left component is a grid heat map.I chose salary paid employees in particular, because only them has annual salary. Then I grouped the dataset by department and job titles to get the sum of annual salary.I also selected the top 100 annual salary as the original data, because the entire dataset has too much dataï¼Œwhich could not be displayed in the heat map.Then I created a pivot table based on the previous "groupby" dataframe. This pivot table could be used as "color" data when I create the heatmap. The right component is a scatter plot.I grouped the dataset by department, job titles, and especiallyï¼Œemployee names (which indicated employee' total annual salary grouped by departments and jobs). Then I picked a smaple dataset for the sample scatter plot. For the linking step, I draged out the "department" and "job titles" column qas a list to give them a specific number to indicate changes. When we click on a grid, its coordinate( x represents job titles number, y represents department number) could be used to update the total annual salary value as well as the "Annual salary - Name" plot.By looking at the distribution of employee salaries (the salary level of most employees, the number of high- or low-paid employees, etc.), we could analyze whether the salary distribution of each department and job title is reasonable.
